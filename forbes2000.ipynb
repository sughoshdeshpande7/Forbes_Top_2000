```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("Forbes2000.csv",encoding='ISO-8859-1')
```

```python
df
```

```python
input_cols = list(df.columns)[1:-1]
target_col = 'marketvalue'
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = df.select_dtypes('object').columns.tolist()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean').fit(df[numeric_cols])
df[numeric_cols] = imputer.transform(df[numeric_cols])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(df[numeric_cols])
df[numeric_cols] = scaler.transform(df[numeric_cols])
```

```python
imputer_cat = SimpleImputer(strategy = 'most_frequent').fit(df[categorical_cols])
df[categorical_cols] = imputer_cat.transform(df[categorical_cols])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df[categorical_cols] = df[categorical_cols].apply(lambda col: le.fit_transform(col))
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[input_cols],df[target_col], test_size=0.25, random_state=1)
```
